```python
import yfinance as yf

msft = yf.Ticker("MSFT")

# ^ returns a named tuple of Ticker objects

fb = yf.Ticker("FB")
shop =yf.Ticker("SHOP")
BYND=yf.Ticker("BYND")
NFLX=yf.Ticker("NFLX")
PINS=yf.Ticker("PINS")
SQ=yf.Ticker("SQ")
TTD=yf.Ticker("TTD")
OKTA=yf.Ticker("OKTA")
SNAP= yf.Ticker("SNAP")
DDOG=yf.Ticker("DDOG")
```

```python
msftdata= msft.history(start="2020-05-14",end="2020-05-15",interval="1m")
```

```python
msftdata.to_csv("msft.csv")
msftdata
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession \
        .builder \
        .appName("Analysis") \
        .getOrCreate()
```

```python
msft_df = spark.read.csv('msft.csv', header=True, inferSchema=True, encoding='utf-8')
```

```python
msft_df.show(5)
```

```python
msft_df.createOrReplaceTempView("msft")
msft_sql= spark.sql("select high,low,Datetime as ts ,'msft' as name from msft ")
msft_sql.show(5)
```

```python
msft_sql.write.json("msft.json")
```

```python
df_json = spark.read.json("*.json")
```

```python
df_json_whithColumn=df_json.withColumn("hours",df_json["ts"][12:2]).drop("low","ts")
df_json_whithColumn.show(10)
```

```python
df_group_hours=df_json_whithColumn.groupBy("name","hours").max().orderBy("name","hours")
df_group_hours.show(10)
```

```python
import pandas as pd
```

```python
pd_df=df_group_hours.toPandas()
```
